```python
dict_clearing = {}

dict_clearing['name'] = []
dict_clearing['spaces'] = []
dict_clearing['denizens'] = []
#dict_clearing['n_exits'] = []
dict_clearing['exits_to'] = []

dict_clearing['marquese_presence'] = []
dict_clearing['dinasty_presence'] = []
dict_clearing['forest_alliance_presence'] = []
dict_clearing['cultists_presence'] = []
dict_clearing['company_presence'] = []
dict_clearing['corvid_presence'] = []
dict_clearing['ducquies_presence'] = []
dict_clearing['lord_presence'] = []
dict_clearing['iron_presence'] = []

dict_clearing['marquese_percent'] = []
dict_clearing['dinasty_percent'] = []
dict_clearing['forest_alliance_percent'] = []
dict_clearing['cultists_percent'] = []
dict_clearing['company_percent'] = []
dict_clearing['corvid_percent'] = []
dict_clearing['ducquies_percent'] = []
dict_clearing['lord_percent'] = []
dict_clearing['iron_percent'] = []

#river and ruins 30% each
dict_clearing['river'] = []
dict_clearing['ruins'] = []

#population_intention
dict_clearing['population_intention'] = []

for i in range (0,12):
    dict_clearing['name'].append(names.human().split()[-1])
    dict_clearing['spaces'].append(randint(1,3))
    
    
    dict_clearing['ruins'].append(True if randint(1,4) == 1 else False)
    
    # exits
    n_exits = randrange(2,5)
    list_exits = []
    
    # not repeated or itself    
    while len(list_exits) < n_exits:   
        exit_number = randint(0,11)
        if exit_number != i and exit_number not in list_exits:
            list_exits.append(exit_number)
    
    #river
    river = True if randint(1,4) == 1 else False
    dict_clearing['river'].append(river)
    n_river = randint(1,2)
    while (river == True) and (len(list_exits) < (n_river + n_exits)):   
        exit_number = randint(0,11)
        if exit_number != i and exit_number not in list_exits:
            list_exits.append(str(exit_number) + ' (river)')
        
    dict_clearing['exits_to'].append(list_exits)
    
    
    #denizens
    denizens_perc = randint(0,100)
    if denizens_perc<=30:
        dict_clearing['denizens'].append('Rabbits')
    elif denizens_perc<=60:
        dict_clearing['denizens'].append('Mouses')
    elif denizens_perc<=90:
        dict_clearing['denizens'].append('Foxes')
    else:
        dict_clearing['denizens'].append('Birds')
    
    population_intention = randint(0,10)
    if population_intention<=1:
        dict_clearing['population_intention'].append('Marquese')
    elif population_intention<=2:
        dict_clearing['population_intention'].append('Dinasty')
    elif population_intention<=3:
        dict_clearing['population_intention'].append('Forest Alliance')
    elif population_intention<=4:
        dict_clearing['population_intention'].append('Cultists')
    elif population_intention<=5:
        dict_clearing['population_intention'].append('Company')
    elif population_intention<=6:
        dict_clearing['population_intention'].append('Corvid')
    elif population_intention<=7:
        dict_clearing['population_intention'].append('Ducquies')
    elif population_intention<=8:
        dict_clearing['population_intention'].append('Lord')
    elif population_intention<=9:
        dict_clearing['population_intention'].append('Iron')
    else:
        dict_clearing['population_intention'].append('None')
    
    dict_clearing['marquese_percent'].append(80)
    dict_clearing['dinasty_percent'].append(20)
    dict_clearing['forest_alliance_percent'].append(30)
    dict_clearing['cultists_percent'].append(30)
    dict_clearing['company_percent'].append(10)
    dict_clearing['corvid_percent'].append(40)
    dict_clearing['ducquies_percent'].append(20)
    dict_clearing['lord_percent'].append(20)
    dict_clearing['iron_percent'].append(30)
    
    dict_clearing['marquese_presence'].append(True if randint(0,100)<=80 else False)
    dict_clearing['dinasty_presence'].append(True if randint(0,100)<=20 else False)
    dict_clearing['forest_alliance_presence'].append(True if randint(0,100)<=30 else False)
    dict_clearing['cultists_presence'].append(True if randint(0,100)<=30 else False)
    dict_clearing['company_presence'].append(True if randint(0,100)<=10 else False)
    dict_clearing['corvid_presence'].append(True if randint(0,100)<=40 else False)
    dict_clearing['ducquies_presence'].append(True if randint(0,100)<=20 else False)
    dict_clearing['lord_presence'].append(True if randint(0,100)<=20 else False)
    dict_clearing['iron_presence'].append(True if randint(0,100)<=30 else False)

#dict_clearing
```

```python
df_clearing = pd.DataFrame.from_dict(dict_clearing)

df_clearing.marquese_percent = df_clearing.apply(lambda x: x['marquese_percent'] + 10 if not x['marquese_presence'] else 20, axis=1)
df_clearing.dinasty_percent = df_clearing.apply(lambda x: x['dinasty_percent'] + 20 if not x['dinasty_presence'] else 10, axis=1)
df_clearing.forest_alliance_percent = df_clearing.apply(lambda x: x['forest_alliance_percent'] + 5 if not x['forest_alliance_presence'] else 0, axis=1)
df_clearing.cultists_percent = df_clearing.apply(lambda x: x['cultists_percent'] + 15 if not x['cultists_presence'] else 15, axis=1)
df_clearing.company_percent = df_clearing.apply(lambda x: x['company_percent'] + 5 if not x['company_presence'] else 0, axis=1)
df_clearing.corvid_percent = df_clearing.apply(lambda x: x['corvid_percent'] + 15 if not x['corvid_presence'] else 15, axis=1)
df_clearing.ducquies_percent = df_clearing.apply(lambda x: x['ducquies_percent'] + 10 if not x['ducquies_presence'] else 10, axis=1)
df_clearing.lord_percent = df_clearing.apply(lambda x: x['lord_percent'] + 20 if not x['lord_presence'] else 20, axis=1)
df_clearing.iron_percent = df_clearing.apply(lambda x: x['iron_percent'] + 10 if not x['iron_presence'] else 15, axis=1)

#soldiers
df_clearing['marquese_force'] = df_clearing.marquese_presence.apply(lambda x: randint(0,8) if x else False)
df_clearing['dinasty_force'] = df_clearing.dinasty_presence.apply(lambda x: randint(0,8) if x else False)
df_clearing['forest_alliance_force'] = df_clearing.forest_alliance_presence.apply(lambda x: randint(0,4) if x else False)
df_clearing['cultists_force'] = df_clearing.cultists_presence.apply(lambda x:  randint(0,5) if x else False)
df_clearing['company_force'] = df_clearing.company_presence.apply(lambda x:  randint(1,8) if x else False)
df_clearing['corvid_force'] = df_clearing.corvid_presence.apply(lambda x:  randint(0,3) if x else False)
df_clearing['ducquies_force'] = df_clearing.ducquies_presence.apply(lambda x:  randint(1,8) if x else False)
df_clearing['lord_force'] = df_clearing.lord_presence.apply(lambda x:  randint(1,6) if x else False)
df_clearing['iron_force'] = df_clearing.iron_presence.apply(lambda x:  randint(1,6) if x else False)

df_clearing
```

```python
list_column = {'occupied_spaces': [[] for _ in  df_clearing]} 
df_clearing['occupied_spaces'] = pd.DataFrame.from_dict(list_column) # create column with lists


def update_occupied_spaces(row, limit):
    row_building = []
    row_token = []
    n = 0
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append('sawmill [ ]')
        n +=1    
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append('recruiter [ ]')
        n += 1    
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append('workshop [ ]')
        n += 1
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append('sawmill [ ]')
        n +=1   
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append('recruiter [ ]')
        n += 1    
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append('workshop [ ]')
        n += 1
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append('sawmill [ ]')
        n +=1    
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append('recruiter [ ]')
        n += 1    
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append('workshop [ ]')
        n += 1
    if row['marquese_presence'] and randint(0, 100) <= 20:
        row_building.append(str(randint(1,5)) + 'wood ( )')
    
    if row['dinasty_presence'] and randint(0, 100) <= 50:
        row_building.append('nest [ ]')
        n += 1
    
    if row['forest_alliance_presence'] and randint(0, 100) <= 60:
        row_token.append('sympathy ( )')
    
    if row['forest_alliance_presence'] and randint(0, 100) <= 15:
        row_building.append('base [ ]')
        n += 1
    
    if row['cultists_presence'] and randint(0, 100) <= 30:
        row_building.append('garden [ ]')
        n += 1
    if row['cultists_presence'] and randint(0, 100) <= 30:
        row_building.append('garden [ ]')
        n += 1
    if row['cultists_presence'] and randint(0, 100) <= 30:
        row_building.append('garden [ ]')
        n += 1
    
    if row['company_presence'] and randint(0, 100) <= 20:
        row_token.append('trading post ( )')
    
    if row['corvid_presence'] and randint(0, 100) <= 70:
        row_token.append('plot ( )')
    
    if row['ducquies_presence'] and randint(0, 100) <= 30:
        row_token.append('hole ( )')
    
    if row['ducquies_presence'] and randint(0, 100) <= 20:
        row_building.append('citadel [ ]')
        n += 1
    
    if row['ducquies_presence'] and randint(0, 100) <= 20:
        row_building.append('market [ ]')
        n += 1
    if row['ducquies_presence'] and randint(0, 100) <= 20:
        row_building.append('citadel [ ]')
        n += 1   
    if row['ducquies_presence'] and randint(0, 100) <= 20:
        row_building.append('market [ ]')
        n += 1
    if row['ducquies_presence'] and randint(0, 100) <= 20:
        row_building.append('citadel [ ]')
        n += 1    
    if row['ducquies_presence'] and randint(0, 100) <= 20:
        row_building.append('market [ ]')
        n += 1
    if row['ducquies_presence'] and randint(0, 100) <= 20:
        row_building.append('citadel [ ]')
        n += 1

    if row['lord_presence'] and randint(0, 100) <= 50:
        row_token.append('mod ( )')
    
    if row['lord_presence'] and randint(0, 100) <= 15:
        row_building.append('stronghold [ ]')
        n += 1
    if row['lord_presence'] and randint(0, 100) <= 15:
        row_building.append('stronghold [ ]')
        n += 1
    if row['lord_presence'] and randint(0, 100) <= 15:
        row_building.append('stronghold [ ]')
        n += 1
    
    if row['iron_presence'] and randint(0, 100) <= 70:
        row_building.append('waystation [ ]')
        n += 1
    
    if row['iron_presence'] and randint(0, 100) <= 10:
        row_token.append('relic ( )')
    
    if n > limit:
        quant = []
        for i in range(0,limit):
            quant.append(row_building[randint(0,n-1)])
        
        row_final = quant + row_token 
    else:
        row_final = row_token + row_building
    return row_final
    
#occupied spaces
df_clearing['occupied_spaces'] = df_clearing.apply(lambda row: update_occupied_spaces(row, row['spaces']), axis=1)
```
